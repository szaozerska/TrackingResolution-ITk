```python
import numpy as np
from math import *
import matplotlib.pyplot as plt

from trackingerror import plot_fixedp, plot_fixedeta
from ITk import ITk
from ATLAS import ATLAS_PhaseII
```

```python
itk = ITk()
atlas = ATLAS_PhaseII()

plot_fixedp(atlas, 'sigma(d)', B=2, p=2, input_mode='detector', label='ATLAS Phase II, p=1 GeV, mass=0.106')
plot_fixedp(itk, 'sigma(d)', B=2, p=2, input_mode='detector', label='ITk, p=1 GeV, mass=0.106')
```

```python
plot_fixedp(atlas, 'sigma(z)', B=2, p=2, input_mode='detector', label='ATLAS Phase II, p=2 GeV, mass=0.106')
plot_fixedp(itk, 'sigma(z)', B=2, p=2, input_mode='detector', label='ITk, p=2 GeV, mass=0.106')
```

```python
# RATIO COMPARISON: SIGMA(D)
# (could have been turned into a function, but I'll leave as is for now)

eta = np.linspace(0, 2.0, 100)
results_itk = []
results_id3 = []
var = 'sigma(d)'
for i in range(len(eta)):
    results_itk.append(itk.errorcalculation(p=2, B=2, eta=eta[i], m=0.106)[var])
    results_id3.append(atlas.errorcalculation(p=2, B=2, eta=eta[i], m=0.106)[var])
ratio = np.array(results_itk) / np.array(results_id3)
plt.plot(eta, ratio)
plt.xlabel('$\eta$')
plt.ylabel('{}: ITk / ATLAS Phase II'.format(var))
```

(Doesn't quite match the results in https://atlas.web.cern.ch/Atlas/GROUPS/PHYSICS/PAPERS/IDTR-2023-01/ so far, but it might be so because of the different detector layout params)
